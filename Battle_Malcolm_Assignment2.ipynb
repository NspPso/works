```python
# Missing Values Imputation
import numpy as np
import pandas as pd
from sklearn.base import TransformerMixin

class DataFrameImputer(TransformerMixin):
    def __init__(self):
        #Impute missing values.
    
        self.fill = None  # Stores imputed values

    def fit(self, X, y=None):
        """Find the most frequent value for categorical and mean for numerical."""
        self.fill = pd.Series(
            [
                X[c].mode()[0] if X[c].dtype == object else X[c].mean()
                for c in X
            ], 
            index=X.columns
        )
        return self

    def transform(self, X, y=None):
        """Fill missing values using stored mode/mean values."""
        return X.fillna(self.fill)

# Apply the class to the data set
df_Malware = DataFrameImputer().fit_transform(df_Malware)

# Verify missing values are gone
print("Total Missing Values After Imputation:", df_Malware.isnull().sum().sum())
```

```python
import warnings
from sklearn.model_selection import train_test_split

# Suppress deprecation warnings I should of done this from the beginning in the first line of the code like german credit. Lots of red warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# Define the target column
target_column = "HasDetections"  # Change if needed

# Define attributes (exclude the target column)
attributes = [col for col in df_Malware.columns if col != target_column]

# Convert sparse data types to normal float64 (if necessary)
df_Malware = df_Malware.astype("float64")

# Split into 80% training and 20% testing
train_x, test_x, train_y, test_y = train_test_split(df_Malware[attributes], df_Malware[target_column], 
                                                    test_size=0.2, random_state=123)

# Print sizes
print(f"Training Features: {train_x.shape}, Training Labels: {train_y.shape}")
print(f"Testing Features: {test_x.shape}, Testing Labels: {test_y.shape}")

```

```python
print(test_y.shape)
```

```python
from sklearn import tree
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score

# Train Decision Tree Classifier
clf = tree.DecisionTreeClassifier()
clf.fit(train_x, train_y)

# Make Predictions
pred_y = clf.predict(test_x)

# Evaluate Model
print("Decision Tree Performance:")
print(classification_report(test_y, pred_y))

# Print 
accuracy = accuracy_score(test_y, pred_y)
precision = precision_score(test_y, pred_y, average='binary')
recall = recall_score(test_y, pred_y, average='binary')
f1 = f1_score(test_y, pred_y, average='binary')

print(f"\nAccuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
```
