```python
# we will focus on categories, genres, and platforms because out of these columns because they have multiple values in the cells which we seen above.
# first were going to make dummy variables for each unique category like we did in class for credit "A14, A13 A12...." so were doing some hot- encoding

# Clean_list_function. I did not know existed. converts list in to a usable string and esures the data is handled properly

def clean_list_column(column):
     return column.apply(lambda x: x.strip("[]").replace("'", "").split(", ") if isinstance(x, str) else x)


#apply celaning to the mult-value columns

df['genres'] = clean_list_column(df['genres'])
df['platforms'] = clean_list_column(df['platforms'])
df['categories'] = clean_list_column(df['categories'])


# converting list into seperate binary columns so when this goes through Machine learning it understands because it dosent understand strings
genres_encoded = df['genres'].explode().str.get_dummies().groupby(level=0).sum()
platforms_encoded = df['platforms'].explode().str.get_dummies().groupby(level=0).sum()
categories_encoded = df['categories'].explode().str.get_dummies().groupby(level=0).sum()

#merge the dummy columns within the DataFrame
columns_to_drop = ['genres', 'platforms', 'categories']
df = df.drop(columns=[col for col in columns_to_drop if col in df.columns])

#display the udpated DataFrame
print(df.head())
             

     

```
